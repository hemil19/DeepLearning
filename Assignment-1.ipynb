```python
#A
import pandas as pd
import numpy as np
#B
headers=["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
 "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
 "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
 "peak-rpm","city-mpg","highway-mpg","price"]
df=pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv',names=headers)
#C
df.head()
```

```python
#D There are illegal '?' character in the dataframe
```

```python
#E
for i in df.columns:
    df[i]=df[i].replace('?',np.nan)
df.head()
```

```python
#F
df.isnull().sum()[df.isnull().sum()>0]
```

```python
#G
price_null_index=df.index[df['price'].isnull()==True]
for i in price_null_index:
    df.drop(i,inplace=True)
```

```python
df.head(15)
```

```python
df.isnull().sum()[df.isnull().sum()>0]
```

```python
#H
columns=['normalized-losses','bore','stroke','horsepower','peak-rpm']
for i in columns:
    df[i]=df[i].astype('float64')
    df[i].fillna(df[i].mean(),inplace=True)
```

```python
#I
df['num-of-doors'].fillna(df['num-of-doors'].mode()[0],inplace=True)
df.head()
```

```python
#J
df.reset_index(inplace=True)
```

```python
df.isnull().sum()[df.isnull().sum()>0]
```

```python
#K
df.dtypes
```

```python
df['price']=df['price'].astype(int)
```

```python
#L
df['city-L/100km']=235/df['city-mpg']
df['highway-L/100km']=235/df['highway-mpg']
```

```python
#M
df['length']=df['length']/df['length'].max()
df['width']=df['width']/df['width'].max()
df['height']=df['height']/df['height'].max()
```

```python
#N
import matplotlib.pyplot as plt
%matplotlib inline
plt.hist(df['horsepower'])
```

```python
#O
labels = ['low', 'medium', 'high'] 
df['horsepower-binned'] = pd.qcut(df['horsepower'], q=3, labels = labels)
df['horsepower-binned'].value_counts().plot(kind='bar')
```
