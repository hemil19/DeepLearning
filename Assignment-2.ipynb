```python
#23
from scipy.stats import pearsonr
corr, _ = pearsonr(dataframe['wheel-base'],dataframe['price'])
print('Pearsons correlation: %.3f' % corr)
print('P value: %.3f' % _)
"""
Null hypothesis is rejected as p-val<0.05
"""
```

```python
#24
from scipy.stats import f_oneway
f_oneway(dataframe['price'][dataframe['drive-wheels'] == '4wd'],
               dataframe['price'][dataframe['drive-wheels'] == 'fwd'],
               dataframe['price'][dataframe['drive-wheels'] == 'rwd'])
"""
F-statistic=67.95 suggests the between-groups variance is 67x the within-group variance
p-value is <0.05, so we reject null hypothesis i.e, both are correlated and the result is significant .
"""
```
